[Back to Contents](#Contents)

### Optimal weights trained by tensorflow for 2:1:2
logs: <br>

```python
def sig(z):
    return 1/(1+np.exp(-z))
```

```python
# One hot encoding of the output **YOU ONLY NEED TO RUN THIS ONCE**
Output = keras.utils.to_categorical(Output,num_classes=2)
```

```python
W1 = np.array([[ 9.52738476],[9.52739143]])
W2 = np.array([[-5.03374052,5.03291368]])
b1 = np.array([ 9.64117527])
b2 = np.array([ 4.6423893,-4.64157534])

min_weights = -30
max_weights = 30
num_of_points = 1000

hidden_nodes = 1
W1 += min_weights 
W2 += min_weights
b1 += min_weights
b2 += min_weights

all_costs = []
counter = 0
for i in np.linspace(min_weights,max_weights,num_of_points):
    W1 += abs(max_weights - min_weights)/num_of_points
    W2 += abs(max_weights - min_weights)/num_of_points
    b1 += abs(max_weights - min_weights)/num_of_points
    b2 += abs(max_weights - min_weights)/num_of_points    
    
    h1 = sig(np.dot(Input,W1) + b1)
    h2 = sig(np.dot(h1,W2) + b2)
    cost = ((Output - h2)**2).mean()
    if counter % 100 ==0:
        print("============================== {} ==============================".format(i))
#         print("W1: ",W1)
#         print("W2: ",W2)
        print("Cost: ",cost)
    
    all_costs.append(cost)
    counter+= 1
plt.plot(np.linspace(min_weights,max_weights,num_of_points), all_costs)
plt.show()
```

```python
all_costs_1 = all_costs
```
