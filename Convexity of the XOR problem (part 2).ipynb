```python
W1 = np.array([[-0.27494025],[ 0.12869418]])
W2 = np.array([[8.19297129e-05,8.32365622e-05]])
```

```python
def sig(z):
    return 1/(1+np.exp(-z))
```

```python
# One hot encoding of the output **YOU ONLY NEED TO RUN THIS ONCE**
Output = keras.utils.to_categorical(Output,num_classes=2)
```

```python
W1 = np.array([[-0.27494025],[ 0.12869418]])
W2 = np.array([[8.19297129e-05,8.32365622e-05]])

# W1 = np.array([[ 0.60528243],[-0.09560066]])
# W2 = np.array([[5.01825753e-07,8.10786446e-07]])

min_weights = -30
max_weights = 30
num_of_points = 1000

hidden_nodes = 1
W1 += min_weights 
W2 += min_weights

all_costs = []

for i in np.linspace(min_weights,max_weights,num_of_points):
#     print(i)
    W1 += abs(max_weights - min_weights)/num_of_points
    W2 += abs(max_weights - min_weights)/num_of_points
    
    h1 = sig(np.dot(Input,W1))
    h2 = sig(np.dot(h1,W2))
    cost = ((Output - h2)**2).mean()
    if i % 100 ==0:
        print("============================== {} ==============================".format(i))
        print("W1: ",W1)
        print("W2: ",W2)
        print("Cost: ",cost)
    
    all_costs.append(cost)

plt.plot(np.linspace(min_weights,max_weights,num_of_points), all_costs)
plt.legend()
plt.show()
```
